## Gradient Descent

```python
##########################
##########################

## TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################

# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(new_x) ## [REQUIRED]
    new_x =new_x - step_multiplier * gradient  ## [REQUIRED]

    step_size = abs(new_x - previous_x)
    # print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))## [REQUIRED])

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
```

## Gradient Descent

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

## Minimise $$f(x, y) = \frac{1}{3^{-x^2 - y^2} + 1}$$

Minimise $$f(x, y) = \frac{1}{r + 1}$$ where $r$ is $3^{-x^2 - y^2}$

```python
## TASK-8 : Code the same things as they were done for the first two examples

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function and its gradient
def f1(x, y):
    return 13 - x**2 - y**2 + 1

def df1(x, y):
    return np.array([-2 * x, -2 * y])

def gradient_descent_f1(initial_guess, multiplier=0.1, precision=0.0001, max_iter=300):
    x, y = initial_guess
    x_list = [x]
    y_list = [y]
    cost_list = [f1(x, y)]

    for _ in range(max_iter):
        gradient = df1(x, y)
        x, y = x - multiplier * gradient[0], y - multiplier * gradient[1]

        step_size = np.linalg.norm(gradient)
        if step_size < precision:
            break

        x_list.append(x)
        y_list.append(y)
        cost_list.append(f1(x, y))

    return x, y, x_list, y_list, cost_list

# Run gradient descent
initial_guess_f1 = (2, 2)
x_min, y_min, x_list_f1, y_list_f1, cost_list_f1 = gradient_descent_f1(initial_guess_f1)

# Plot the function and gradient descent path
x = np.linspace(-4, 4, 100)
y = np.linspace(-4, 4, 100)
X, Y = np.meshgrid(x, y)
Z = f1(X, Y)

plt.figure(figsize=[15, 5])

# Plot cost function
ax1 = plt.subplot(1, 2, 1, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax1.scatter(x_list_f1, y_list_f1, [f1(x, y) for x, y in zip(x_list_f1, y_list_f1)], color='red')
ax1.set_title('Function f1(x, y)')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('f(x, y)')

# Plot cost reduction
plt.subplot(1, 2, 2)
plt.plot(cost_list_f1, color='blue')
plt.title('Cost Reduction (f1)')
plt.xlabel('Iteration')
plt.ylabel('Cost')

plt.tight_layout()
plt.show()

```

```python
# Define the function and its gradient
def f2(x, y):
    r = 3 - x**2 - y**2
    return 1 / (r + 1)

def df2(x, y):
    r = 3 - x**2 - y**2
    grad_x = 2 * x / (r + 1)**2
    grad_y = 2 * y / (r + 1)**2
    return np.array([grad_x, grad_y])

def gradient_descent_f2(initial_guess, multiplier=0.1, precision=0.0001, max_iter=300):
    x, y = initial_guess
    x_list = [x]
    y_list = [y]
    cost_list = [f2(x, y)]

    for _ in range(max_iter):
        gradient = df2(x, y)
        x, y = x - multiplier * gradient[0], y - multiplier * gradient[1]

        step_size = np.linalg.norm(gradient)
        if step_size < precision:
            break

        x_list.append(x)
        y_list.append(y)
        cost_list.append(f2(x, y))

    return x, y, x_list, y_list, cost_list

# Run gradient descent
initial_guess_f2 = (1, 1)
x_min, y_min, x_list_f2, y_list_f2, cost_list_f2 = gradient_descent_f2(initial_guess_f2)

# Plot the function and gradient descent path
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = f2(X, Y)

plt.figure(figsize=[15, 5])

# Plot cost function
ax2 = plt.subplot(1, 2, 1, projection='3d')
ax2.plot_surface(X, Y, Z, cmap='plasma', edgecolor='none')
ax2.scatter(x_list_f2, y_list_f2, [f2(x, y) for x, y in zip(x_list_f2, y_list_f2)], color='red')
ax2.set_title('Function f2(x, y)')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('f(x, y)')

# Plot cost reduction
plt.subplot(1, 2, 2)
plt.plot(cost_list_f2, color='red')
plt.title('Cost Reduction (f2)')
plt.xlabel('Iteration')
plt.ylabel('Cost')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

def mse(y, y_hat):
    """
    Calculate the Mean Squared Error between the true values and the predicted values.

    Parameters:
    y (numpy array): Array of true values.
    y_hat (numpy array): Array of predicted values.

    Returns:
    float: Mean Squared Error.
    """
    mse_calc = np.mean((y - y_hat) ** 2)
    return mse_calc

# Example usage
# Assuming y and y_hat are defined numpy arrays
y = np.array([1, 2, 3, 4, 5])
y_hat = np.array([0.8, 2.2, 2.9, 3.6, 5.1])

# Calculate MSE
mse_value = mse(y, y_hat)
print(f"The Mean Squared Error is: {mse_value}")
```

```python
import numpy as np

# Function to compute gradients
def grad(x, y, thetas):
    n = y.size
    theta0_slope = (-2 / n) * np.sum(y - (thetas[0] + thetas[1] * x))
    theta1_slope = (-2 / n) * np.sum((y - (thetas[0] + thetas[1] * x)) * x)
    return np.array([theta0_slope, theta1_slope])

# Example data
x_5 = np.array([1, 2, 3, 4, 5])
y_5 = np.array([2, 4, 6, 8, 10])
thetas = np.array([0.5, 1.5])  # Initial theta values
multiplier = 0.01

# Gradient Descent Loop
for i in range(1000):
    gradient = grad(x_5, y_5, thetas)
    thetas = thetas - multiplier * gradient

    # Optionally print the current theta values or cost to monitor progress
    if i % 100 == 0:
        print(f"Iteration {i}: Thetas: {thetas}")

print("Final thetas:", thetas)
```
